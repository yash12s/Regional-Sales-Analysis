### **EDA**

```python
df['order_month'] = df['order_date'].dt.to_period('M')

monthly_sales = df.groupby('order_month')['revenue'].sum()

plt.figure(figsize=(15,4))

monthly_sales.plot(marker='o',color='navy')

from matplotlib.ticker import FuncFormatter
formatter = FuncFormatter(lambda x, pos: f'{x/1e6:.1f}M')
plt.gca().yaxis.set_major_formatter(formatter)

plt.title('Monthly Sales Trend')
plt.xlabel('Month')
plt.ylabel('total Revenue (Millions)')

plt.xticks(rotation=45)

plt.tight_layout()
plt.show()


```

```python

df['order_month_num'] = df['order_date'].dt.month
df['order_month_name'] = df['order_date'].dt.strftime('%b')

monthly_seasonality = df.groupby(['order_month_num', 'order_month_name'])['revenue'].sum().reset_index()

monthly_seasonality = monthly_seasonality.sort_values('order_month_num')

plt.figure(figsize=(12, 5))
plt.plot(monthly_seasonality['order_month_name'], monthly_seasonality['revenue'], marker='o', color='teal')

formatter = FuncFormatter(lambda x, _: f'{x/1e6:.1f}M')
plt.gca().yaxis.set_major_formatter(formatter)

plt.title('Overall Monthly Sales Seasonality (All Years Combined)', fontsize=16)
plt.xlabel('Month', fontsize=12)
plt.ylabel('Total Revenue (Millions)', fontsize=12)
plt.grid(True)
plt.tight_layout()


plt.show()
```

```python
# get top 10 products by revenue
top_products = df.groupby('product_name')['revenue'].sum().sort_values(ascending=False).head(10)

plt.figure(figsize=(12, 6))
ax = sns.barplot(
    x=top_products.values,
    y=top_products.index,
    palette='viridis'
)

for i, (value, name) in enumerate(zip(top_products.values, top_products.index)):
    ax.text(value + 1e5, i, f'{value/1e6:.2f}M', va='center', fontsize=10)

plt.title('Top 10 Products by Revenue', fontsize=16)
plt.xlabel('Total Revenue')
plt.ylabel('Product Name')

formatter = FuncFormatter(lambda x, _: f'{x/1e6:.1f}M')
ax.xaxis.set_major_formatter(formatter)

plt.tight_layout()
plt.show()
```

```python
# get bottom 10 products by revenue
bottom_products = df.groupby('product_name')['revenue'].sum().sort_values(ascending=True).head(10)

plt.figure(figsize=(12, 6))
ax = sns.barplot(
    x=bottom_products.values,
    y=bottom_products.index,
    palette='Reds_r'
)

for i, (value, name) in enumerate(zip(bottom_products.values, bottom_products.index)):
    ax.text(value + 1e4, i, f'{value/1e6:.2f}M', va='center', fontsize=10)

plt.title('Bottom 10 Products by Revenue', fontsize=16)
plt.xlabel('Total Revenue')
plt.ylabel('Product Name')

formatter = FuncFormatter(lambda x, _: f'{x/1e6:.1f}M')
ax.xaxis.set_major_formatter(formatter)

plt.tight_layout()
plt.show()
```

```python
### Sales by Channel

chan_sales = df.groupby('channel')['revenue'].sum().sort_values(ascending=True)

plt.figure(figsize=(5,5))

plt.pie(
    chan_sales.values,
    labels=chan_sales.index,
    autopct='%1.1f%%',
    colors=sns.color_palette('coolwarm')
)

plt.title('Total Sales by Channel')

plt.tight_layout()

plt.show()
```

```python
### Average Order Value(AOV) Distribution

aov = df.groupby('order_number')['revenue'].sum()

plt.figure(figsize=(12,4))

plt.hist(
    aov,
    bins=50,
    color='lightgreen',
    edgecolor='black'
)

plt.title('Distribution of Average Order Value')
plt.xlabel('Order Value (USD)')
plt.ylabel('Number of Orders')

plt.tight_layout()

plt.show()
```

```python
### Unit Price Distribution per Product

plt.figure(figsize=(16,6))

product_order = df.groupby('product_name')['unit_price'].mean().sort_values(ascending=False).index

sns.boxplot(
    data=df,
    x='product_name',
    y='unit_price',
    order=product_order,
    palette='viridis'
)

plt.title('Unit Price Distribution per Product')
plt.xlabel('Product Name')
plt.ylabel('Unit Price (USD)')
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

```python
### Top 10 states by revenue and Order count

state_stats = df.groupby('state_name').agg({
    'revenue': 'sum',
    'order_number': 'nunique'  # Count of unique orders
}).rename(columns={'order_number': 'order_count'})

top_states = state_stats.sort_values('revenue', ascending=False).head(10)

plt.figure(figsize=(12,6))
sns.set_style("whitegrid")


ax1 = sns.barplot(
    x=top_states.index,
    y=top_states['revenue'],
    color='royalblue',
    label='Revenue'
)

ax2 = ax1.twinx()

sns.barplot(
    x=top_states.index,
    y=top_states['order_count'],
    color='salmon',
    alpha=0.6,
    label='Order Count',
    ax=ax2
)

ax1.set_ylabel('Revenue (USD)', color='royalblue')
ax2.set_ylabel('Order Count', color='salmon')
ax1.set_xlabel('State')
plt.title('Top 10 States by Revenue and Order Count')
ax1.set_xticklabels(ax1.get_xticklabels(), rotation=45)

h1, l1 = ax1.get_legend_handles_labels()
h2, l2 = ax2.get_legend_handles_labels()
ax1.legend(h1 + h2, l1 + l2, loc='upper left')

plt.tight_layout()
plt.show()
```

```python
### Average Profit Marging by Channel

avg_margin_channel = df.groupby('channel')['profit_margin_pct'].mean().sort_values(ascending=False)

plt.figure(figsize=(8,5))
sns.set_style("whitegrid")

ax = sns.barplot(
    x=avg_margin_channel.index,
    y=avg_margin_channel.values,
    palette='viridis'
)

for i, v in enumerate(avg_margin_channel.values):
    ax.text(i, v + 1, f'{v:.1f}%', ha='center', va='bottom', fontweight='bold')

plt.title('Average Profit Margin (%) by Channel')
plt.ylabel('Profit Margin (%)')
plt.xlabel('Channel')
plt.tight_layout()
plt.show()
```

```python
### top and bottom 10 customers by revenue
customer_revenue = df.groupby('customer_name')['revenue'].sum()

top10_customers = customer_revenue.sort_values(ascending=False).head(10)
bottom10_customers = customer_revenue.sort_values().head(10)

fig, axes = plt.subplots(1, 2, figsize=(16, 6))
sns.set_style("whitegrid")

# Top 10
sns.barplot(
    x=top10_customers.values,
    y=top10_customers.index,
    palette='Greens_d',
    ax=axes[0]
)
axes[0].set_title('Top 10 Customers by Revenue')
axes[0].set_xlabel('Revenue (USD)')
axes[0].set_ylabel('Customer')

# Bottom 10
sns.barplot(
    x=bottom10_customers.values,
    y=bottom10_customers.index,
    palette='Reds_d',
    ax=axes[1]
)
axes[1].set_title('Bottom 10 Customers by Revenue')
axes[1].set_xlabel('Revenue (USD)')
axes[1].set_ylabel('')

plt.tight_layout()
plt.show()
```

```python
### customer segmentation revenue vs profit margin

cust_seg = df.groupby('customer_name').agg({
    'revenue': 'sum',
    'profit_margin_pct': 'mean'
}).reset_index()

plt.figure(figsize=(10, 6))
sns.set_style("whitegrid")

sns.scatterplot(
    data=cust_seg,
    x='revenue',
    y='profit_margin_pct',
    s=80,
    alpha=0.7,
    color='royalblue'
)

plt.title('Customer Segmentation: Revenue vs. Profit Margin (%)')
plt.xlabel('Total Revenue per Customer (USD)')
plt.ylabel('Average Profit Margin (%)')

# Optional: Add horizontal and vertical lines for median split
plt.axvline(cust_seg['revenue'].median(), color='gray', linestyle='--', linewidth=1)
plt.axhline(cust_seg['profit_margin_pct'].median(), color='gray', linestyle='--', linewidth=1)

plt.tight_layout()
plt.show()
```

```python
### Correlation Heatmap

numeric_df = df.select_dtypes(include=['float64', 'int64'])

corr_matrix = numeric_df.corr()

# Plot heatmap
plt.figure(figsize=(10, 6))
sns.heatmap(
    corr_matrix,
    annot=True,
    fmt='.2f',
    cmap='coolwarm',
    square=True,
    linewidths=0.5,
    cbar_kws={"shrink": .8}
)

plt.title('Correlation Heatmap of Numeric Features')
plt.tight_layout()
plt.show()
```
